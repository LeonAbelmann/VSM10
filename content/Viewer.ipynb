```python
import piplite
await piplite.install("numpy")
await piplite.install("matplotlib") 
await piplite.install("ipywidgets") 

import ipywidgets as widgets
from IPython.display import display

import vsm
import importlib
importlib.reload(vsm)

uploader = widgets.FileUpload(accept='.vhd', multiple=False)
out = widgets.Output()

def analyze(_):
    with out:
        out.clear_output()

        if not uploader.value:
            print("No file uploaded yet.")
            return

        file_info = uploader.value[0]
        fname = file_info['name']
        data = file_info['content']

        # save file into notebook filesystem
        with open(fname, "wb") as f:
            f.write(data)

        # read first line
        with open(fname, "r") as f:
            first_line = f.readline().strip()

        print("Uploaded:", fname)
        print("First line:", first_line)

        # your normal analysis
        print("before plotVSM")
        Results = vsm.plotVSM(fname)
        print("after plot VSM")
        Results.print()

button = widgets.Button(description="Analyze VHD")
button.on_click(analyze)

display(uploader, button, out)
```
