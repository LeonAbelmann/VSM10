# VSM Data Analyzer
1. Click the "Run" button (>) to start the interface.
2. Upload a `.vhd` file using the "Upload" button.
3. Click "Analyze VHD" to process the file.

```python
# imports
import piplite
await piplite.install("numpy")
await piplite.install("matplotlib") 
await piplite.install("ipywidgets") 

import ipywidgets as widgets
from IPython.display import display

import vsm
import VSMPlot
import importlib
importlib.reload(VSMPlot)

# Analysis and plot setttings
Settings = vsm.VSMSettings()

# Setup buttons
uploader = widgets.FileUpload(accept='.vhd', multiple=False)
out = widgets.Output()

def analyze(_):
    with out:
        out.clear_output()

        if not uploader.value:
            print("No file uploaded yet.")
            return

        file_info = uploader.value[0]
        fname = file_info['name']
        data = file_info['content']

        # save file into notebook filesystem
        with open(fname, "wb") as f:
            f.write(data)

        # read first line
        with open(fname, "r") as f:
            first_line = f.readline().strip()

        print("Uploaded:", fname)
        # print("First line:", first_line)

        # analyze and plot uploaded file
        Results = VSMPlot.plotVSM(fname,Settings)
        Results.print()

button = widgets.Button(description="Analyze VHD")
button.on_click(analyze)

display(uploader, button, out)
```
